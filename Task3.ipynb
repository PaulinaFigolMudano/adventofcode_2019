```python
import pandas as pd
import requests
import os
import re
import matplotlib.pyplot as plt
import turtle

f = open("data3.txt", "r")
numbers = f.read()
```

```python
numbers[0:100]
```

```python
split = numbers.split('\n')
wire_1 = split[0]
wire_2 = split[1]
```

```python
split2 = wire_1.split(',')
split2

new_list = list([[0,0]])

for c in range(len(split2)):
    if re.findall(r"\b[A-Z]+", split2[c])[0][0] == 'R':
        old = new_list[c]
        new_list.append([old[0] + int(split2[c][1:]), old[1]])
    elif re.findall(r"\b[A-Z]+", split2[c])[0][0] == 'L':
        old = new_list[c]
        new_list.append([old[0] - int(split2[c][1:]), old[1]])
    if re.findall(r"\b[A-Z]+", split2[c])[0][0] == 'D':
        old = new_list[c]
        new_list.append([old[0],old[1] - int(split2[c][1:])])
    if re.findall(r"\b[A-Z]+", split2[c])[0][0] == 'U':
        old = new_list[c]
        new_list.append([old[0],old[1] + int(split2[c][1:])]) 
        
new_list        
```

```python
from matplotlib import pyplot as plt
from shapely.geometry.polygon import LinearRing, Polygon, LineString

ring = LineString(new_list)
x, y = ring.xy

fig = plt.figure(1, figsize=(5,5), dpi=90)
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_title('Polygon Edges')

xrange = [-1000, 9500]
yrange = [-1000, 9500]
ax.set_xlim(*xrange)
ax.set_ylim(*yrange)
ax.set_aspect(1)
```

```python
split3 = wire_2.split(',')
split3

new_list2 = list([[0,0]])

for c in range(len(split3)):
    if re.findall(r"\b[A-Z]+", split3[c])[0][0] == 'R':
        old = new_list2[c]
        new_list2.append([old[0] + int(split3[c][1:]), old[1]])
    elif re.findall(r"\b[A-Z]+", split3[c])[0][0] == 'L':
        old = new_list2[c]
        new_list2.append([old[0] - int(split3[c][1:]), old[1]])
    if re.findall(r"\b[A-Z]+", split3[c])[0][0] == 'D':
        old = new_list2[c]
        new_list2.append([old[0],old[1] - int(split3[c][1:])])
    if re.findall(r"\b[A-Z]+", split3[c])[0][0] == 'U':
        old = new_list2[c]
        new_list2.append([old[0],old[1] + int(split3[c][1:])]) 
```

```python
from matplotlib import pyplot
from shapely.geometry import LineString
#from figures import SIZE

BLUE =   '#6699cc'
YELLOW = '#ffcc33'
GREEN =  '#339933'
GRAY =   '#999999'

def plot_coords(ax, ob):
    x, y = ob.xy
    ax.plot(x, y, 'o', color=GRAY, zorder=1)

fig = pyplot.figure(1, figsize=(9,9), dpi=90) #1, figsize=(10, 4), dpi=180)

a = LineString(new_list)
b = LineString(new_list2)

# 1: disconnected multilinestring
ax = fig.add_subplot(111)

plot_coords(ax, a)

plot_coords(ax, b)

x, y = a.xy
ax.plot(x, y, color=YELLOW, alpha=0.5, linewidth=3, solid_capstyle='round', zorder=2)

x, y = b.xy
ax.plot(x, y, color=GREEN, alpha=0.5, linewidth=3, solid_capstyle='round', zorder=2)

#ax.set_title('a) lines')

xrange = [-5000, 9500]
yrange = [-5000, 9500]
ax.set_xlim(*xrange)
#ax.set_xticks(range(*xrange) + [xrange[-1]])
ax.set_ylim(*yrange)
#ax.set_yticks(range(*yrange) + [yrange[-1]])
ax.set_aspect(1)

#2: invalid self-touching ring
ax = fig.add_subplot(111)

x, y = a.xy
ax.plot(x, y, color=GRAY, alpha=0.7, linewidth=1, solid_capstyle='round', zorder=1)
x, y = b.xy
ax.plot(x, y, color=GRAY, alpha=0.7, linewidth=1, solid_capstyle='round', zorder=1)

for ob in a.intersection(b):
    x, y = ob.xy
    if len(x) == 1:
        ax.plot(x, y, 'o', color=BLUE, zorder=2)
    else:
        ax.plot(x, y, color=BLUE, alpha=0.7, linewidth=3, solid_capstyle='round', zorder=2)

ax.set_title('b) collection')

xrange = [-5000, 9500]
yrange = [-5000, 9500]
ax.set_xlim(*xrange)
#ax.set_xticks(range(*xrange) + [xrange[-1]])
ax.set_ylim(*yrange)
#ax.set_yticks(range(*yrange) + [yrange[-1]])
ax.set_aspect(1)

pyplot.show()
```
