```python
import pandas as pd
import requests
import os
import re

f = open("data2.txt", "r")
numbers = f.read()
```

```python
numbers_int = [int(s) for s in re.findall(r'\b\d+\b', numbers)]
numbers_int[0:10]
```

```python
#before running the program, replace position 1 with the value 12 and 
#replace position 2 with the value 2. What value is left at position 0 after the program halts?
numbers_int[1] = 12
numbers_int[2] = 2
```

```python
#create a sequence:
start = 0
stop  = len(numbers_int)-1
step  = 4
# to get inclusive range change stop as stop+step
stop +=step #now stop is 12

list_seq = list()
for i in range(start, stop, step):
    list_seq.append(i)
    
list_seq = list_seq[:-1]
```

```python
for i in list_seq:
    index1 = numbers_int[i+1]
    index2 = numbers_int[i+2]
    index3 = numbers_int[i+3]
    if numbers_int[i] == 1:  
        addition = numbers_int[index1] + numbers_int[index2]
        numbers_int[index3] = addition
    elif numbers_int[i] == 2:
        multiplication = numbers_int[index1] * numbers_int[index2]
        numbers_int[index3] = multiplication
    elif numbers_int[i] == 99:
        break
numbers_int             
```
